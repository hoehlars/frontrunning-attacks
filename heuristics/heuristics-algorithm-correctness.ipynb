### Evaluating correctness of heuristics algorithm

To evaluate the correctness of the heuristcs algorithm, we randomly sampled rows from the insertion-attack.csv and check if the algorithm detects the same attacker 1, victim and attacker 2.

```python
# Import packages
from web3 import Web3
import pandas as pd
from tqdm import tqdm

# Import Heuristics Algorithm
from application.backend.heuristics.InsertionAttackHeuristicsDetector import InsertionAttackHeuristicsDetector
```

```python
web3 = Web3(Web3.HTTPProvider("https://intensive-sly-mountain.quiknode.pro/a3f5256d7f2af6541d483cce3f1d49c94c01879e/"))
insertionAttackDetector = InsertionAttackHeuristicsDetector(web3)
```

First load Insertion attack csv:

```python
df_insertion = pd.read_csv ('../data/insertion_attacks.csv', delimiter=',')
print('# of insertion: ', len(df_insertion))
```

Sample 1000 rows:

```python
df_insertion_sampled_1000 = df_insertion.sample(n=1000)
df_insertion_sampled_1000
```

```python
def is_heuristics_similar_as_insertion_attack_csv(row_insertion_attacks, result_heuristics_algorithm):
    
    # Extract attackers & victim from sampled row of insertion csv
    insertion_csv_row_address_attacker_1 = row_insertion_attacks["First Attacker"]
    insertion_csv_row_address_victim = row_insertion_attacks["Whale"]
    insertion_csv_row_address_attacker_2 = row_insertion_attacks["Second Attacker"]
    
    # Create a mask for rows where the specified columns have the target values
    mask = ((result_heuristics_algorithm['first_attacker'] == insertion_csv_row_address_attacker_1) &
            (result_heuristics_algorithm['victim'] == insertion_csv_row_address_victim) &
            (result_heuristics_algorithm['second_attacker'] == insertion_csv_row_address_attacker_2))

    # Filter the DataFrame
    matching_rows = result_heuristics_algorithm[mask]
    
    return not matching_rows.empty

    
    
```

```python
not_detected_heuristics = []
amount_detected_heuristics = 0

for index, row in tqdm(df_insertion_sampled_1000.iterrows(), total=df_insertion_sampled_1000.shape[0], desc="Processing rows"):
    
    block_number = row["Block Number"]
    
    try:
        # Run heuristics algorithm on Block Number
        result_heuristics_algorithm = insertionAttackDetector.get_frontrunning_attacks_of_block(block_number)
    except:
        continue
    
    if result_heuristics_algorithm.empty:
        not_detected_heuristics.append(block_number)
        continue
    
    heuristics_similar_as_insertion_attack_csv = is_heuristics_similar_as_insertion_attack_csv(row, result_heuristics_algorithm)
    
    if heuristics_similar_as_insertion_attack_csv:
        amount_detected_heuristics += 1
    else:
        not_detected_heuristics.append(block_number)
        
print(not_detected_heuristics)
print(amount_detected_heuristics)
```

Calculate the percentage of randomly sampled insertion attacks that were detected by the heuristics algorithm.

```python
(amount_detected_heuristics - len(not_detected_heuristics)) / amount_detected_heuristics
```
